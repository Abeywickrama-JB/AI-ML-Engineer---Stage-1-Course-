```python
import matplotlib.pyplot as plt
import numpy as np

# value for x
x = np.linspace(-10, 10, 100)

# y = x^2
y = x**2

# Drawing Graph
plt.figure(figsize=(8, 5))
plt.plot(x, y, label="y = x^2", color="blue", linewidth=2)

plt.title("Plot of the Function y = x^2")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.legend()

# output
plt.show()
```
