```python
# b) Program to find max, min, and average of 10 student marks

marks = []

print("Enter marks for 10 students:")
for i in range(1, 11):
    mark = float(input(f"Enter mark for student {i}: "))
    marks.append(mark)

# Calculation
maximum_mark = max(marks)
minimum_mark = min(marks)
average_mark = sum(marks) / len(marks)

#  Print
print("\n--- Results ---")
print(f"Maximum Mark: {maximum_mark}")
print(f"Minimum Mark: {minimum_mark}")
print(f"Average Mark: {average_mark:.2f}")
```

```python
# c) Program to input 5 marks and display the grade

total_marks = 0

print("Enter 5 marks for the student:")
for i in range(1, 6):
    mark = float(input(f"Enter mark {i}: "))
    total_marks += mark

# Calculation
average_mark = total_marks / 5
print(f"\nAverage Mark: {average_mark:.2f}")

# Conditional Statements
if average_mark > 75:
    grade = "A"
elif 65 <= average_mark <= 75:
    grade = "B"
elif 55 <= average_mark <= 64:
    grade = "C"
elif 45 <= average_mark <= 54:
    grade = "S"
else:
    grade = "F"

print(f"Final Grade: {grade}")
```

```python
# d) Program to sum numbers until -999 is entered

total_sum = 0
print("Enter numbers to sum (Enter -999 to stop):")

while True:
    number = float(input("Enter a number: "))

    # -999 loop is stop
    if number == -999:
        break

    total_sum += number

print(f"\nThe total sum of the entered numbers is: {total_sum}")
```
