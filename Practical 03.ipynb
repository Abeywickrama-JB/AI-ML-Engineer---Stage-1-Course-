```python
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# ==========================================
# Step 1: Read the data in CSV file
df = pd.read_csv('Mobile_Price_Classification-220531-204702.csv')

# Assumption: The last column is 'price_range' (0 or 1) and the others are Features
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values

# Scaling the features to the same level (this is important for an ANN to be successful)
scaler = StandardScaler()
X = scaler.fit_transform(X)

# ==========================================
# Step 2: Split into Training (75%) and Testing (25%)
# ==========================================
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# ==========================================
# Step 3: ANN Model Architecture
# ==========================================
model = tf.keras.models.Sequential([
    # 1st Hidden Layer including the Input Layer (8 neurons)
    tf.keras.layers.Dense(units=8, activation='relu', input_shape=(X_train.shape[1],)),

    # 2nd Hidden Layer (4 neurons)
    tf.keras.layers.Dense(units=4, activation='relu'),

    # Output Layer (1 neuron - Binary Classification )
    tf.keras.layers.Dense(units=1, activation='sigmoid')
])

# ==========================================
# Step 4: Compile the model
# ==========================================
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

# ==========================================
# Step 5: Fit and Train the model (100 Epochs, Batch Size 32)
# ==========================================
print("Training the ANN Model...")
history = model.fit(X_train, y_train, epochs=100, batch_size=32, validation_data=(X_test, y_test))

# ==========================================
# Step 6: Save the weights file
# ==========================================
model.save_weights('mobile_price_ann_weights.weights.h5')
print("\nModel weights saved successfully as 'mobile_price_ann_weights.weights.h5'")
```
